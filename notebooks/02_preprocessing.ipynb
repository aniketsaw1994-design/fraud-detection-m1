## Preprocessing overview:
I will be performing data preprocessing to handle different issues within the data and preparing the data for my baseline model which in this case is Logistic Regression.
I am choosing logistic regression because I want to check how the features are performing and if the issue can be solved with this model first.
Columns that I will processing are:
1. Amount: The Amount column is on a completely different scale from the V1-V28 features. Those V features are PCA outputs already centered around zero with small ranges, while Amount runs from 0 to 25,691. Logistic regression, my baseline model, is gradient-based and sensitive to feature scale and a feature with a large magnitude like Amount would dominate the model's coefficient updates simply because of its size, not its actual importance. Standardization rescales Amount to have a mean of 0 and standard deviation of 1, bringing it onto the same footing as the V features so the baseline model treats all features fairly. I chose standardization over min-max normalization because the V1-V28 features are already standardized-like (centered near zero from PCA), so standardizing Amount makes it consistent with them. The tree-based models that follow are unaffected by this scaling either way, so standardizing Amount is safe for them and beneficial for the baseline.
2. Time: I will be droping this column because this column does not specifically indicates time of the day but rather time elapsed from the first transaction and the actual time of the first transaciton is unknown. Although I do consider that time can be a crucial factor in identifying the fraud and I have shown that in EDA where there was a spike in fraudulent transactions when legitimate transactions dropped. But this Time column does not specify anything so I will be dropping it.
3. V1-V28: I will be keeping all of these columns for now, if my Logistic regression model does not work out well and I move to complex tree-based model then these tree-based models find signal in combinations of features, even ones with weak individual correlation. A feature with near-zero correlation to the target alone can become highly predictive in combination with others.
4. Sampling: For Sampling the data I will be using 80-20 Stratified split because our data is highling imbalanced and startified split deliberately preserves the original class proportion in both sets. If fraud is 0.17% of the whole dataset, it remains 0.17% in both train and test.
5. Scaling fit: The scaler learns two values from the data the mean and the standard deviation which it then uses to transform every value. If I calculated those values from the entire dataset before splitting, information from the test set (its mean and spread) would leak into the training process. The model would then have indirectly "seen" characteristics of the test data during training, making the evaluation dishonest and overly optimistic.
To prevent this data leakage, I fit the scaler on the training set only, learning the mean and standard deviation from training data alone. I then apply that same fitted scaler to transform the test set. This mirrors production reality when a new transaction arrives: You scale it using statistics learned from past data, because you obviously can't use future data you don't yet have.
